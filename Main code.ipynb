```python
#display all missing values in the dataset
missing_values = data.isnull().sum()
print(missing_values)
```

```python
sampled_data = data.sample(frac=0.1, random_state=42)

# Display the count of missing values before imputation
print("Missing Values Before Imputation:")
print(sampled_data.isnull().sum())

# Impute missing values (replace this part with your specific imputation logic)
imputed_data = sampled_data.fillna(sampled_data.mean())

# Display the count of missing values after imputation
print("\nMissing Values After Imputation:")
print(imputed_data.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
print("Missing Values in 'fuel_consumption' Before Imputation:")
print(sampled_data['fuel_consumption_l_100km'].isnull().sum())
#changing the categorical columns to numerical using label encoder
label_encoder = LabelEncoder()
sampled_data['fuel_type_encoded'] = label_encoder.fit_transform(sampled_data['fuel_type'])
sampled_data['color_encoded'] = label_encoder.fit_transform(sampled_data['color'])
sampled_data['power_kw_encoded'] = label_encoder.fit_transform(sampled_data['power_kw'])
sampled_data['power_ps_encoded'] = label_encoder.fit_transform(sampled_data['power_ps'])
sampled_data['price_in_euro_encoded'] = label_encoder.fit_transform(sampled_data['price_in_euro'])

# Impute missing values for the encoded column
sampled_data['fuel_type_encoded'] = sampled_data['fuel_type_encoded'].fillna(sampled_data['fuel_type_encoded'].mean())
sampled_data['color_encoded'] = sampled_data['color_encoded'].fillna(sampled_data['color_encoded'].mean())
sampled_data['power_kw_encoded'] = sampled_data['power_kw_encoded'].fillna(sampled_data['power_kw_encoded'].mean())
sampled_data['power_ps_encoded'] = sampled_data['power_ps_encoded'].fillna(sampled_data['power_ps_encoded'].mean())
sampled_data['price_in_euro_encoded'] = sampled_data['price_in_euro_encoded'].fillna(sampled_data['price_in_euro_encoded'].mean())
# Display the count of missing values after imputation
print("\nMissing Values in 'fuel_consumption' After Imputation:")
print(sampled_data['fuel_type_encoded'].isnull().sum())
print(sampled_data['color_encoded'].isnull().sum())
print(sampled_data['power_kw_encoded'].isnull().sum())
print(sampled_data['power_ps_encoded'].isnull().sum())
```

```python
from sklearn.neighbors import LocalOutlierFactor
features = ['price_in_euro_encoded', 'power_kw_encoded']
X = sampled_data[features]

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)  # Adjust parameters as needed
outlier_scores = lof.fit_predict(X)

# Display the count of outliers detected by LOF
print("\nOutliers Detected by LOF:")
print(np.sum(outlier_scores == -1))

# Display the scatter plot with colors indicating outliers
plt.scatter(range(len(sampled_data)), sampled_data['price_in_euro'], c=outlier_scores, cmap='viridis')
plt.xlabel('Data Point Index')
plt.ylabel('Price')
plt.title('Outliers Detection with LOF (price_in_euro and power_kw)')
plt.colorbar()
plt.show()
```

```python
#how does the type of fuel impact the pricing of used cars. Is there a notable difference in price distribution?
#using the sample data represents 10% of the dataset to execute easily
plt.figure(figsize=(12, 8))
sns.boxplot(x='fuel_type_encoded', y='price_in_euro_encoded', data=sampled_data)
plt.title('Impact of Fuel Type on Used Car Prices')
plt.xlabel('Fuel Type')
plt.ylabel('Price in Euro')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
